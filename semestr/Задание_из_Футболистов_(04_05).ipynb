```python
import pandas as pd
data=('transfusion.data')
df=pd.read_csv(data)
df.head()
```

```python
df.columns
```

```python
df.info ()
```

```python
df.rename({'Recency (months)' : 'Частота донации'}, axis='columns', inplace=True)
df.rename({'Frequency (times)' : 'Количество раз'}, axis='columns', inplace=True)
df.rename({'Monetary (c.c. blood)' : 'Стоимость'}, axis='columns', inplace=True)
df.rename({'Time (months)' : 'Месяцев донорства'}, axis='columns', inplace=True)
df.rename({'whether he/she donated blood in March 2007' : 'Сдача крови в марте 2007'}, axis='columns', inplace=True)
df.columns
```

```python
grouped = df.groupby('Частота донации', as_index=False)['Стоимость'].mean()
# добавим сортировку по убыванию
grouped.sort_values(by='Стоимость', ascending=False)
```

```python
df.groupby('Частота донации')['Стоимость'].agg(['mean', 'min', 'max', 'median'])
```

```python
df.groupby(['Частота донации', 'Количество раз'], as_index=False)['Стоимость'].mean()
```

```python
df.groupby(['Частота донации', 'Количество раз']).apply(lambda x: x.sort_values(by='Стоимость', ascending=False))
```

```python
df.groupby(['Частота донации', 'Количество раз']).apply(lambda x: x.sort_values(by='Стоимость', ascending=False)).\
                                                    drop(['Частота донации', 'Количество раз'], axis=1)
```

```python
print('Среднее:', round(df['Количество раз'].mean(), 2),
      'Медиана:', df['Количество раз'].median(),
      'Мода:', df['Количество раз'].mode()[0])
```
