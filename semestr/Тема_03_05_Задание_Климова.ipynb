## Задание 1.

Прочитать данные
`https://raw.githubusercontent.com/yakushinav/omo/main/data/movies.csv`

Имена столбцов
```
['color', 'director_name', 'num_critic_for_reviews', 'duration',
                'gross', 'movie_title', 'num_user_for_reviews', 'country',
                'cotent_rating', 'budget', 'title_year', 'imdb_score', 'genre']
```

````python
import pandas as pd
url1 = 'https://raw.githubusercontent.com/yakushinav/omo/main/data/movies.csv'
df1 = pd.read_csv(url1,
                 header=None,
                 sep= '|',
                 names= ['color', 'director_name', 'num_critic_for_reviews', 'duration',
                'gross', 'movie_title', 'num_user_for_reviews', 'country',
                'cotent_rating', 'budget', 'title_year', 'imdb_score', 'genre'],
                 na_values=['', '?', '-'],)
df1.head()
````
